# ML类别：
- （1）监督学习：有特征，有标签的数据。这里面还涉及到一个 自监督学习：标号是直接从数据里面生成出来的，如bert和Transformer等

- （2）半监督学习：有一些有标签的数据，还有大量没有标签的数据。这里面涉及到两个任务：一个是像监督学习一样学习一个模型来预测标号，也尽量利用那些没有标号好的信息；另一个是将没有标号的数据，给预测出标号。比如自学习：通过少量的有标签的数据训练模型，然后将没标签的数据送入模型，得到预测标签，将置信度高的标签数据，再次加入训练集，进行下一轮的模型训练，重复这个过程。
- （3）无监督学习：整个数据集没有标号，任务不是预测一个标号。比如一个任务是聚类，即把数据根据相似性聚成几个类；另一类任务是Density Estimation（即密度估计），也就是说将数据的分布估计出来，如生成对抗模型。
- （4）强化学习：一个模型和环境交互，从环境中获取一些观察点，进行学习，之后再对环境做出反应，然后得到环境对模型的反馈，再进行学习，这样重复循环。比如机器人走路。






# 监督学习模型训练

监督学习模型训练一共四个最重要的事：
- 1、模型：有输入和输出；
- 2、损失：衡量预测值与真实值的误差
- 3、目标：解决任务的目标，如目标是使得预测值更接近真实值
- 4、优化：最小化损失函数，也就是调整可学习参数



# 监督学习模型可分为以下四大类：
- 1、决策树：用一些树做一些决定，如XGB、LGBM等；
- 2、线性模型：线性组合；
- 3、核方法：使用核函数来衡量两个样本的特征之间的一些相似度，这样就可以达到非线性的效果；
- 4、神经网络：多层的神经网络来提取特征，最后接一个线性层来预测结果。

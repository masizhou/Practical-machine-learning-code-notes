# 线性模型：

线性模型简单，但是面试中也有可能被问住，下面回答一个挺绕的问题：回归和分类的损失函数。

- 我们都知道，使用线性模型做回归，那么模型预测一个值，都有一个真实的标签与之对应，因此，我们使用MSE来作为损失函数，来不断地降低损失函数值。

- 但是，如果使用线性模型做分类，依然使用MSE作为损失是不对的，为啥？因为：假设我们现在使用MSE作为二分类的损失了，那么每一个预测的类别（概率，是二维的，比如到A的概率和到B的概率）和真实的标签，做MSE，会发现无论是A，还是B都在不断地降低损失，这是不对的。因为我们只需要预测属于某一个类别，也就是说降低其中的一个类别的损失，而最大化其他类别的损失。因此需要使用交叉熵损失函数。

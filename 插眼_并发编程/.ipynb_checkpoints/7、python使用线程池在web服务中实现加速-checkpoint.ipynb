# 3、代码：

```python
import flask
import json
import time
from concurrent.futures import ThreadPoolExecutor

app = flask.Flask(__name__)
pool = ThreadPoolExecutor()


def read_file(): # 模仿读文件io
    time.sleep(0.1)
    return "file result"


def read_db(): # 模仿读数据库io
    time.sleep(0.2)
    return "db result"


def read_api(): # 模仿读api io
    time.sleep(0.3)
    return "api result"


@app.route("/")
def index():
    result_file = pool.submit(read_file) # 使用pool.submit()方法实现线程池的加速，下面同理
    result_db = pool.submit(read_db)
    result_api = pool.submit(read_api)

    return json.dumps({
        "result_file": result_file.result(), # 要获取结果，直接.result()即可，下面同理
        "result_db": result_db.result(),
        "result_api": result_api.result(),
    })


if __name__ == "__main__":
    app.run()
```

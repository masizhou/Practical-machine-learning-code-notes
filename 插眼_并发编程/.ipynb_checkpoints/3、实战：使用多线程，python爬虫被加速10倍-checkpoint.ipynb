# 2、案例：

```python
import threading
import time
import requests

urls = [
    f"https://www.cnblogs.com/#p{page}"
    for page in range(1, 50 + 1)
]


def craw(url): # 爬取网页
    r = requests.get(url)
    print(url, len(r.text), r.status_code)


# craw(urls[2])


def single_thread():
    print("single_thread begin")
    for url in urls: # 使用for循环，一页一页爬取
        craw(url)
    print("single_thread end")


def multi_thread():
    print("multi_thread begin")
    threads = []把所有的url先加入到线程中
    for url in urls: # 
        threads.append(
            threading.Thread(target=craw, args=(url,))
        )

    for thread in threads: # 开始多线程
        thread.start()

    for thread in threads: # 结束每个线程
        thread.join()

    print("multi_thread end")


if __name__ == "__main__":
    start = time.time()
    single_thread()
    end = time.time()
    print("single thread cost:", end - start, "seconds")

    start = time.time()
    multi_thread()
    end = time.time()
    print("multi thread cost:", end - start, "seconds")
```

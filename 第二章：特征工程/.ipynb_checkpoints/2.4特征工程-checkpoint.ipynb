# 特征工程
特侦工程是机器学习中最重要的一步，传统的机器学习算法通常需要我们自己手动的提取特征，所以特征工程做的好了，最终的结果很大程度上就是好的，反之亦然。但是人为的提取特征不但具有认知上的误差，而且工作量极大，因此后来出现的深度学习才会大放光彩。具体来说，深度学习和传统机器学习的不同之处在于：不需要人为的提取特征，这一步可以让网络自己学习，进而提取特征。具体如下图所示：
